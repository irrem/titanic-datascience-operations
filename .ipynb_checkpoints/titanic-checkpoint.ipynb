18.Please create a pie chart that shows the percentage of people who survived according to the their age group. For the age group, we will have [0-18] as child, [19-60] as adult, [60 and over] as old.

```python
import pandas as pd 
data = pd.read_csv("train.csv") 
data.dropna()
bins= [0,18,60,81]
labels = ['Children','Adult','Old']
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
data[['AgeGroup', 'Survived']].groupby(['AgeGroup'], as_index=True).mean().sort_values(by='AgeGroup', ascending=True)
#data['AgeGroup'].value_counts().plot(kind='pie',autopct='%1.1f%%',title="Age Groups In Titanic Survived")
data[data.Survived==1].AgeGroup.value_counts().plot(kind='pie',autopct='%1.1f%%',title="Age Groups In Titanic Survived")
#data.AgeGroup.value_counts()
#print(data.Age[data.Age>=60].count())
```
