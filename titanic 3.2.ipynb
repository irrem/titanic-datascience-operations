16.Please create a strip chart (You can use the library seaborn for this) that plots age (y-axis), and survival (x-axis).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
data = pd.read_csv("train.csv") 
data.dropna()
sns.set_theme(style="whitegrid")
ax = sns.stripplot(y = 'Age', hue = 'Survived', palette = 'Set1', data = data)
ax.set(title = 'Passenger status Survival and Age', xlabel = 'Survival', ylabel = 'Age')
plt.show()
```
